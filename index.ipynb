```python
import pandas as pd
```

```python
df = pd.read_json('words.json')
df.rename(columns = {'message':'word'}, inplace = True)

df
```

```python
expandeds = []

for index, row in df.iterrows():
    word = list(row['word'])
    expandeds.append(word)
```

```python
expandeds_df = pd.DataFrame(expandeds, columns=['0_char','1_char','2_char','3_char','4_char',])
expandeds_df['word'] = df['word']

expandeds_df
```
